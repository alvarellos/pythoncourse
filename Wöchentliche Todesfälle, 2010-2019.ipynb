# Wöchentliche Todesfälle, 2010-2024

https://opendata.swiss/de/dataset/wochentliche-todesfalle-2010-20241

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["figure.figsize"] = (20, 10)

url='https://dam-api.bfs.admin.ch/hub/api/dam/assets/31786122/master'

df = pd.read_csv(url, sep=';', encoding='ISO 8859-1', nrows=1042)

df['endend'] = pd.to_datetime(df.endend, dayfirst=True)
df

df['Month'] = df['endend'].dt.month
df['AnzTF_HR']=pd.to_numeric(df['AnzTF_HR'], errors='coerce')

filter_young = df[df.Alter.str.contains("64")]
filter_young

pivot_young = filter_young.pivot_table(index='Jahr', columns='Month', values='AnzTF_HR', aggfunc='sum')
pivot_young
```

```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib
import matplotlib as mpl
import pandas as pd


deaths = pivot_young.values  # numpy array of values

fig, ax = plt.subplots()
im = ax.imshow(deaths)

Jahr = ["2010", "2011", "2012", "2013",
              "2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021", "2022", "2023", "2024", ]
Month = ["Jan", "Feb", "Mar",
           "Abr", "May", "Jun", "July", "Ago", "Sept", "Oct", "Nov", "Dec"]
# Show all ticks and label them with the respective list entries
ax.set_xticks(np.arange(len(Month)), labels=Month)
ax.set_yticks(np.arange(len(Jahr)), labels=Jahr)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(Jahr)):
    for j in range(len(Month)):
        text = ax.text(j, i, deaths[i, j],
                       ha="center", va="center", color="w")

ax.set_title("Deaths Months / Year")
fig.tight_layout()
plt.show()
```
